```python
# changing the columns names to snake format.
# everything will be lowercase and space will be replaced with _

df.columns = df.columns.str.lower() # converting evry capital letter to small letter
df.columns = df.columns.str.replace(" ","_") # replacing every space with underscore
```

```python
df.columns
```

```python
# still there issue with purchase_amount_(usd)
# we will simply rename that 

df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# we are going to group customers into age groups "young adult", "adult", "middle age", "senior"

# concept of binning
bins = [18,25,45,60,120]
labels = ["young adult", "adult", "middle age", "senior"]

df["age_group"] = pd.cut(df["age"], bins = bins, labels = labels)
```

```python
df["age_group"]
```

```python
df["age_group"].unique()
```

```python
df[["age","age_group"]]
```

```python
# creating purchase_frequency_days
# we will be converting the texual frequencies into number (number of days between purchases)

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
} 

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df["purchase_frequency_days"].unique()
```

```python
# let's check of the "discount_applied" and "promo_code_used" contain the same information
(df["discount_applied"] == df["promo_code_used"]).all()
```

```python
# that means they are the same values. we can just remove the promo_code_used column.

df = df.drop("promo_code_used", axis = 1)
```
